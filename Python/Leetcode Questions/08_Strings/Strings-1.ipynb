# 151. Reverse Words in a String

```python
class Solution:
    def reverseWords(self, s: str) -> str:
        # Remove leading and trailing whitespace, and split the string into a list of words
        s = s.strip().split()

        # Reverse the order of the words in the list
        s.reverse()

        # Join the reversed words with a space in between to form a single string
        return ' '.join(s)
```

```python
# Algorithm
# Initialize maxlength to 0 and ans to an empty string.
# Iterate through each character in the input string s.
# For each character, consider it as the center of an odd-length palindrome. 
# Expand outward from the center by moving two pointers (l and r) to the left and right respectively. 
# Check if the characters at s[l] and s[r] are equal. 
# Continue expanding and updating ans and maxlength if the current substring is longer than the previously 
# recorded longest palindrome.
# Repeat step 3, but this time consider the character and its adjacent character as the center of an even-length palindrome. 
# Set l to the current index and r to l+1. Expand outward and update ans and maxlength if a longer palindrome is found.
# After iterating through all characters, ans will contain the longest palindrome substring. 
# Return ans.
# In simple words, the algorithm explores each character in the input string and tries to expand outward to find palindromes. 
# It considers both odd-length and even-length palindromes by using different center points. 
# The algorithm keeps track of the longest palindrome found and returns it as the result.

class Solution:
    def longestPalindrome(self, s: str) -> str:
        n = len(s)  # Get the length of the input string
        maxlength = 0  # Variable to store the length of the longest palindrome found
        ans = ""  # Variable to store the longest palindrome substring found

        for i in range(n):  # Iterate through each character in the string
            l = i  # Left pointer initialized to the current index
            r = i  # Right pointer initialized to the current index

            while l >= 0 and r < n and s[l] == s[r]:  # Expand the substring while characters at l and r are equal
                if (r - l + 1) > maxlength:  # Check if current substring is longer than previous longest palindrome
                    ans = s[l:r + 1]  # Update ans with the current substring
                    maxlength = r - l + 1  # Update maxlength with the length of the current substring
                l -= 1  # Move the left pointer to the left
                r += 1  # Move the right pointer to the right

            l = i  # Reset the left pointer to the current index
            r = i + 1  # Reset the right pointer to the next index

            while l >= 0 and r < n and s[l] == s[r]:  # Expand the substring while characters at l and r are equal
                if (r - l + 1) > maxlength:  # Check if current substring is longer than previous longest palindrome
                    ans = s[l:r + 1]  # Update ans with the current substring
                    maxlength = r - l + 1  # Update maxlength with the length of the current substring
                l -= 1  # Move the left pointer to the left
                r += 1  # Move the right pointer to the right

        return ans  # Return the longest palindrome substring found
```

# 13. Roman to Integer

```python
# Algorithm/Intuition:
    
# 1. Initialize a dictionary `d` to store the mapping of Roman numerals to their corresponding integer values.
# 2. Initialize a variable `ans` to store the final result.
# 3. Iterate through each character `s[i]` in the given input string `s`.
# 4. Check if the current Roman numeral `s[i]` is smaller than the next Roman numeral `s[i+1]`.
#    - If true, subtract the value of `s[i]` from the result `ans`.
#    - If false, add the value of `s[i]` to the result `ans`.
# 5. Return the final result `ans`.

class Solution:
    def romanToInt(self, s: str) -> int:
        d = {'I': 1, 'V': 5, 'X': 10, 'L': 50, 'C': 100, 'D': 500, 'M': 1000}  # Mapping of Roman numerals to integers
        ans = 0  # Initialize the result
        
        for i in range(len(s)):
            if i + 1 < len(s) and d[s[i]] < d[s[i + 1]]:  # Check if current numeral is smaller than the next numeral
                ans -= d[s[i]]  # Subtract current numeral's value from the result
            else:
                ans += d[s[i]]  # Add current numeral's value to the result
        
        return ans  # Return the final result
```

# 8. String to Integer (atoi)

```python
class Solution:
    def myAtoi(self, s: str) -> int:
        # Remove leading and trailing whitespace from the input string
        s = s.strip()

        # Get the length of the modified string
        n = len(s)

        # Initialize variables
        sign = False
        ans = 0

        # Check if the string is empty, return 0
        if not s:
            return 0

        # Check if the first character is an alphabet, return 0
        if s[0].isalpha():
            return 0

        # Initialize index variable
        i = 0

        # Check if the first character is a '+' or '-'
        if s[0] == '-' or s[0] == '+':
            # If it is '-', set the sign flag to True
            if s[0] == '-':
                sign = True
            i += 1

        # Process digits while iterating through the string
        while i < n and s[i].isdigit():
            # Convert the digit to an integer and update the answer
            ans = ans * 10 + int(s[i])
            i += 1

        # Apply the sign to the answer
        if sign:
            ans = -1 * ans

        # Constrain the answer within the range of a 32-bit signed integer
        ans = max(min(ans, 2 ** 31 - 1), -2 ** 31)

        # Return the final answer
        return ans
```

# 14. Longest Common Prefix

```python
class Solution:
    def longestCommonPrefix(self, strs: List[str]) -> str:
        ans = ""  # Initialize the common prefix as an empty string
        for i in range(len(strs[0])):  # Iterate over the characters of the first string
            for s in strs:  # Iterate over all the strings in the input list
                if i == len(s) or s[i] != strs[0][i]:
                    # If the index i is out of bounds for the current string or the character at index i doesn't match the character at index i of the first string
                    return ans  # Return the current common prefix
            ans += strs[0][i]  # Append the character at index i to the common prefix
        return ans  # Return the final common prefix
```

# 686. Repeated String Match

```python
# Approach 1
class Solution:
    def repeatedStringMatch(self, a: str, b: str) -> int:
        # Assign a to t for reference
        t = a

        # Get the lengths of strings a and b
        m = len(b)
        n = len(a)

        # Calculate the maximum number of times a needs to be repeated to contain b
        end = m // n + 2

        # Initialize a count variable
        count = 1

        # Iterate up to end to check if b is a substring of t
        for i in range(end):
            # If b is found in t, return the count
            if b in t:
                return count
            else:
                # If b is not found, append a to t and increment the count
                t += a
                count += 1

        # If b is not found after repeating a, return -1 as specified
        return -1
```

```python
# Algorithm/Intuition:
# 1. Define a hash function that converts a given string to a hash value using the Rabin-Karp algorithm.
# 2. Implement the Rabin-Karp algorithm to find a pattern within a text.
#    - Initialize the pattern hash value and the text hash value using the hash function.
#    - Check if the initial hash values and the substrings match. If they do, return the starting index.
#    - Iterate through the text, updating the text hash value using the rolling hash formula.
#    - Check if the updated hash value and the substrings match. If they do, return the starting index.
# 3. Implement the `repeatedStringMatch` function:
#    - Check if the given strings `a` and `b` are equal. If they are, return 1.
#    - Determine the minimum number of repetitions of string `a` required to cover string `b`.
#    - Create a temporary string by repeating string `a` the determined number of times.
#    - If the length of the temporary string is still less than the length of string `b`, append one more repetition of `a`.
#    - Check if the pattern `b` matches within the repeated string using the Rabin-Karp algorithm.
#    - If a match is found, return the count of repetitions.
#    - Check if the pattern `b` matches within the repeated string plus one additional repetition.
#    - If a match is found, return the count of repetitions plus 1.
#    - If no match is found, return -1.

class Solution:
    def hash_fun(self, text):
        h = 0
        for ch in text:
            h = (h * 31 + ord(ch)) % int(1e6)  # Calculate the hash value using the Rabin-Karp algorithm
        return h

    def rabinkarp(self, text, pattern):
        n = len(text)
        m = len(pattern)
        patternhash = self.hash_fun(pattern)  # Calculate the hash value of the pattern
        texthash = self.hash_fun(text[:m])  # Calculate the hash value of the initial substring

        # Check if initial hash values and substrings match
        if texthash == patternhash and text[:m] == pattern:
            return 0

        for i in range(1, n - m + 1):
            # Update texthash using rolling hash formula
            texthash = (texthash * 31 - ord(text[i - 1]) * pow(31, m, int(1e6)) + ord(text[i + m - 1])) % int(1e6)
            # Check if hash values and substrings match
            if texthash == patternhash and text[i:i + m] == pattern:
                return i

        return -1

    def repeatedStringMatch(self, a: str, b: str) -> int:
        if a == b:
            return 1

        count = (len(b) - 1) // len(a) + 1  # Determine the minimum number of repetitions of string a
        temp = a * count  # Create a temporary string by repeating string a
        if len(temp) < len(b):
            temp += a
            count += 1

        # Check if pattern matches in the repeated string
        if self.rabinkarp(temp, b) != -1:
            return count

        # Check if pattern matches in the repeated string plus one additional repetition
        if self.rabinkarp(temp + a, b) != -1:
            return count + 1

        return -1

 
```

# 692. Top K Frequent Words

```python
# Algorithm/Intuition:
# 1.  Create a frequency hash map using the `Counter` function from the `collections` module 
#     to count the occurrences of each word in the given list.
    
# 2.  Sort the words lexicographically by extracting the keys from the frequency 
#     hash map using the `keys()` method and sorting them with the `sorted()` function.
    
# 3.  Sort the words based on their frequencies in descending order and lexicographically in 
#     ascending order using the `sorted()` function with a custom key function.
    
# 4.  Return the top `k` frequent words by slicing the sorted list.

from typing import List
from collections import Counter

class Solution:
    def topKFrequent(self, words: List[str], k: int) -> List[str]:
        freq_hash_map = Counter(words)  # Count the frequency of words using a hash map
        sort_words = sorted(freq_hash_map.keys())  # Sort the words lexicographically
        
        ans = sorted(sort_words, key=lambda x: (-freq_hash_map[x], x))  # Sort the words based on frequency and lexicographically
        
        return ans[:k]  # Return the top k frequent words
```
