```python
class Node:
    def __init__(self,val:int,next=None) -> None:
        self.val = val
        self.next = next
        
class SingleList:
    def __init__(self) -> None:
        self.head = Node(0)
        self.tail = self.head
        self.size = 0
        
    def from_list(self,arr):
        temp = self.head
        if self.size>0:
            temp = self.tail.next
        for num in arr:
            new_node = Node(num)
            temp.next = new_node
            temp = new_node
            self.size+=1
        self.tail = temp
            
    def print_list(self):
        temp = self.head.next
        while temp.next:
            print(temp.val,end="-> ")
            temp = temp.next
        print(temp.val,end="")
        print()
        
    def add(self,val):
        self.size+=1
        temp = self.tail
        temp.next = Node(val)
        self.tail = temp
        
    def insert(self,index,val):
        assert 0<=index<=self.size, "index should be less than linked list size"
        count = 0
        temp = self.head
        new_node = Node(val)
        while count<index:
            temp = temp.next
            count+=1
        t2 = temp.next
        temp.next = new_node
        new_node.next = t2
        self.size += 1
        
    def delete(self,index):
        count = 0
        temp = self.head
        prev = None
        while count<index:
            prev = temp
            temp = temp.next
            count += 1
        self.size-=1
        if prev is None:
            self.head = self.head.next
        else:
            prev.next = temp.next        
                
    def remove(self,val,multiple=False):
        temp = self.head
        found = False
        index = -1
        prev = None
        while temp:
            if temp.val==val:
                found = True
                if prev is None:
                    self.head = self.head.next
                else:
                    prev.next = temp.next
                if multiple is False:
                    break
            prev = temp
            temp = temp.next
            index+=1
            
        if not found:
            print("val does not exist in list")
    
    def search(self,val):
        temp = self.head
        found = False
        index = -1
        while temp:
            if temp.val==val:
                found = True
                break
            temp = temp.next
            index+=1
        return index if index<self.size else -1
    
if __name__ == "__main__":
    ll = SingleList()
    arr = [1,2,3,5]
    ll.from_list(arr)
    ll.add(51)
    ll.from_list([52,53,89,90])
    ll.add(95)
    ll.from_list([101,102,105])
    ll.insert(13,115)
    ll.insert(0,115)
    ll.insert(8,115)
    # ll.print_list()
    # print(ll.size)
    # ll.delete(16)
    # ll.print_list()    
    # ll.delete(0)
    # ll.delete(14)
    # ll.remove(115,multiple=True)
    ll.print_list()    
    # print(ll.search(150))
    # print(ll.size)
    
    slow = ll.head
    fast = ll.head
    count = 0
    while fast.next and fast.next.next:
        slow = slow.next
        fast = fast.next.next
        count+=1
    print(count*2)
    print(slow.val)
    print(ll.size)
```
