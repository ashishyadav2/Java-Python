```python
class Node:
    def __init__(self,val,prev = None,next = None) -> None:
        self.prev = prev
        self.val = val
        self.next = next
        
    
class DoubleLinkedList:
    def __init__(self) -> None:
        self.head = None
        self.tail = None
        self.size = 0
        self.mapping = {}
        
    def add(self,val):
        new_node = Node(val)
        if self.head is None:
            self.head = new_node
            self.tail = new_node
        else:
            new_node.prev = self.tail
            self.tail.next = new_node
            self.tail = new_node
        self.size+=1
        
    def from_list(self,arr):
        for num in arr:
            self.add(num)
        
    def insert(self,index,val):
        assert -1<=index<=self.size, "Out of index"
        if index==self.size or index==-1:
            self.add(val)
        elif index==0:
            new_node = Node(val)
            new_node.next = self.head
            self.head.prev = new_node
            self.head = new_node
            self.size+=1
        else:            
            counter = 0
            temp = self.head
            while counter<index-1:
                temp = temp.next
                counter+=1            
            new_node = Node(val)
            new_node.next = temp.next
            temp.next = new_node
            new_node.prev = temp
            new_node.next.prev = new_node
            self.size+=1
        
    def remove(self,val,multiple=False):
        temp = self.head
        while temp:
            if val==temp.val:
                if temp==self.head:
                    self.head = temp.next
                    if self.head:
                        self.head.prev = None      
                    else:
                        self.tail = None
                elif temp==self.tail:
                    self.tail = temp.prev
                    if self.tail:
                        self.tail.next = None
                else:
                    temp.next.prev = temp.prev
                    temp.prev.next = temp.next
                self.size-=1
                if multiple is False:
                    break
            temp = temp.next    
            
    def pop(self,index):
        assert -1<=index<self.size, "Index out of bound"
        counter = 0
        temp = self.head            
        if index==0:
            self.head = self.head.next
            if self.head:
                self.head.prev = None
            else:
                self.tail = None
            
        elif index==self.size-1 or index==-1:
            self.tail = self.tail.prev
            self.tail.next = None
        else:
            while counter<index:
                temp = temp.next
                counter+=1
            temp.next.prev = temp.prev
            temp.prev.next = temp.next
            
        self.size-=1
        
    def print_list(self):
        if self.head is None:
            print("List is empty")
        else:    
            temp = self.head
            while temp.next:
                print(f"{temp.val} <-> ",end="")
                temp = temp.next
            print(f"{temp.val}")
            
    def print_reversed(self):
        temp = self.tail
        while temp.prev:
            print(f"{temp.val} <-> ",end="")
            temp = temp.prev
        print(f"{temp.val}")
            
    def to_list(self):
        arr = []
        temp = self.head
        while temp:
            arr.append(temp.val)
            temp = temp.next
        return arr
    
    def get_mappings(self):
        temp = self.head
        while temp:
            self.mapping.update({temp:temp.val})
            temp = temp.next
        return self.mapping
        
if __name__ == "__main__":
    dll = DoubleLinkedList()
    dll.add(65)
    dll.add(13)
    dll.add(14)
    dll.from_list([15,17,25,38,42])
    dll.from_list([65,65,65,65])
    dll.insert(5,26)
    dll.print_list()
    dll.remove(65,multiple=True)
    dll.print_list()
    print("size:", dll.size)
    dll.pop(7)
    dll.print_list()
    print(dll.to_list())
    print(dll.get_mappings())
    # print("Head: ",dll.head.val)
    # print("Tail: ",dll.tail.val)
    print("size:", dll.size)
    # dll.print_reversed()
```
