# Categorial, Dummy variables and one hot encoding

```python
import pandas as pd
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv("homeprices.csv")
```

```python
df.head()
```

```python
# to get dummy variables ie non numerical values, categorical nomial variables
dummies = pd.get_dummies(df.town)
dummies
```

```python
#pd.concat to join 2 dataframes axis ="rows|columns"takes an array as an input
merged = pd.concat([df,dummies],axis="columns")
merged
```

```python
# if variable can be derived from other variable it is adivsed to drop them
# as it may create dummy trap drop atleast one column
#drop to delete column/row
final = merged.drop(['town','west windsor'],axis="columns")
final
```

```python
model = linear_model.LinearRegression()
y = df["price"]
x = final.drop("price",axis='columns')
model.fit(x,y)
```

```python
model.predict([[2800,0,1]])
```

```python
model.predict([[3400,0,0]])
```

```python
#to get accuracy
model.score(x,y)
```

# One Hot Encoding

```python
from sklearn.preprocessing import LabelEncoder
le = LabelEncoder()
```

```python
dfle = df
#it takes label column as an input
dfle.town = le.fit_transform(dfle.town)
dfle
```

```python
# values will give 2D array
x = df[['town','area']].values
x
```

```python
from sklearn.preprocessing import OneHotEncoder
from sklearn.compose import ColumnTransformer
```

```python
# Assuming you have a dataset called 'x' with categorical features in the first column
# In the updated code, we create a ColumnTransformer object that specifies the transformation to be applied to the 
# first column ([0]) using the OneHotEncoder. The remainder='passthrough' argument ensures that 
# the remaining columns in the dataset are passed through without any changes.

# You can then apply the transformation by calling fit_transform on the ColumnTransformer 
# object, passing your dataset x as an argument. The result will be stored back in the x 
# variable, containing the encoded data.

ct = ColumnTransformer(
    transformers=[('encoder', OneHotEncoder(), [0])],
    remainder='passthrough'
)
```

```python
x = ct.fit_transform(x)
x
```

```python
# take all the rows and columns from index 1
x = x[:,1:]
x
```

```python
model.fit(x,y)
```

```python
model.predict([[1,0,2800]])
```

```python
model.predict([[0,1,3400]])
```
