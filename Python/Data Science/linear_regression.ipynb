# Simple Linear Regression

## Regression using single variable

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
df = pd.read_csv("mumbai.csv")
```

```python
#plt.scatter(x-axis,y-axis,color="color_name",marker="how_point_should_look")
plt.xlabel("area")
plt.ylabel("price")
plt.scatter(df.area,df.price,color='red')
plt.show()
```

```python
regr = linear_model.LinearRegression()
regr.fit(df[['area']],df.price) #arguement should be 2d array 
#independent dependent
```

```python
regr.predict([[850]]) #values to be passed. here we are using area so area 
#should be passed if more than one variable that values should be passed
# in correct order
```

```python
regr.coef_ # to get coefficient value
```

```python
regr.intercept_ # to get y intercept
```

## Multi variate linear regression

```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
# Loading data
df = pd.read_csv("mumbai.csv")
```

```python
#Preprocessing
df = df[~(df['type_of_building'] == 'Individual House')]
df["Balcony"] = df["Balcony"].fillna(0)
mode = df["Furnished_status"].mode()
df["Furnished_status"] = df["Furnished_status"].fillna(mode)
df["parking"] = df["parking"].fillna(0)
```

```python
plt.scatter(df["area"],df["price"])
plt.show()
```

```python
# making regression model
reg = linear_model.LinearRegression()
reg.fit(df[['area','Bedrooms','Bathrooms','Balcony','parking']],df["price"])
```

```python
#prediction
reg.predict([[1200,2,2,2,2]])
```
