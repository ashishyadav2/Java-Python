```python
import matplotlib.pyplot as plt
import numpy as np
```

# Simple Plot

```python
x = np.array([0.,6])
y = np.array([2,142])
```

```python
plt.plot(x,y)
plt.show()
```

```python
# to plot without lines
plt.plot(x,y,'o')
```

```python
# default x axis
y = np.array([1,5,9,10])
plt.plot(y)
```

```python
# Markers and color
x = [1,6,4,5]
plt.plot(x,y,marker="<",color="red")
```

```python
# Format Strings fmt
# format marker|line|color
# : is dotted line
# ms = marker size
# mec = markeredgecolor
plt.plot(x,y,'*:r',ms=20,mec='b')
```

```python
plt.plot(x,y,linestyle='dashdot')
plt.show()
# short hand property ls = ':'
# linewidth = number
plt.plot(x,y,ls=':',linewidth='5')
plt.show()
```

```python
# adding mulitple lines
x = [1,2,3,4]
y = [25,15,16,30]
x1 = [10,12,14,50]
y1 = [2,5,6,20]
plt.plot(x,y,color="orange")
plt.plot(x1,y1)
plt.show()
```

```python
# Labels
plt.plot(x1,y1)
plt.xlabel('X label')
plt.ylabel('Y label')
# positioning of title
plt.title('plot title',loc='left')
plt.show()
```

```python
# adding grid lines to plot
plt.plot(x1,y1)
plt.xlabel('X label')
plt.ylabel('Y label')
plt.title('plot title',loc='left')
# axis

plt.grid()
plt.plot(x1,y1)
plt.show()

plt.grid(axis='y')
plt.plot(x1,y1)
plt.show()
    
plt.plot(x1,y1)
plt.grid(axis='x',color="red")
plt.show()
```

```python
# Mulitple plots
# plt.subplot(1, 2, 1)
#the figure has 1 row, 2 columns, and this plot is the first plot.
x = np.array([0, 1, 2, 3])
y = np.array([3, 8, 1, 10])

plt.subplot(2, 1, 1)
plt.plot(x,y)
plt.title("title1")
#plot 2:
x = np.array([0, 1, 2, 3])
y = np.array([10, 20, 30, 40])

plt.subplot(2, 1, 2)
plt.plot(x,y)
plt.title("title2")

plt.suptitle("SuperTitle")
plt.show()
```

# Scatter plots

```python
# Scatter plots
x = np.array([5,7,8,7,2,17,2,9,4,11,12,9,6])
y = np.array([99,86,87,88,111,86,103,87,94,78,77,85,86])
plt.scatter(x, y)

x = np.array([2,2,8,1,15,8,12,9,7,3,11,4,7,14,12])
y = np.array([100,105,84,105,90,99,90,95,94,100,79,112,91,80,85])
plt.scatter(x, y)
plt.show()
```

```python
x = np.random.randint(100, size=(100))
y = np.random.randint(100, size=(100))
colors = np.random.randint(100, size=(100))
sizes = 10 * np.random.randint(100, size=(100))

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='nipy_spectral')

plt.colorbar()

plt.show()
```

# Bars

```python
# vertical bars
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y)
plt.show()

#horizontal bars
plt.barh(x,y,color="orange")
plt.show()
```

```python
# bar width for vertical bars
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.bar(x,y,width=0.2)
plt.show()

# bar height for horizontal bars
x = np.array(["A", "B", "C", "D"])
y = np.array([3, 8, 1, 10])

plt.barh(x,y,height=0.2)
plt.show()
```

# Histograms

```python
# used for frequency distributions
x = np.random.normal(170, 10, 250)
plt.hist(x)
plt.show()
```

```python
# Piecharts
y = np.array([35, 25, 25, 15])

plt.pie(y)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]

plt.pie(y, labels = mylabels, startangle = 180)
plt.show() 
```

```python
y = np.array([35, 25, 25, 15])
mylabels = ["Apples", "Bananas", "Cherries", "Dates"]
myexplode = [0.2, 0, 0, 0]

plt.pie(y, labels = mylabels, explode = myexplode,shadow=True)
plt.legend(title = "Fruits")
plt.show() 
```
